```python
all_df = pd.read_csv('data/all_df_nlp_preprocessed.csv')
all_df.head()
```

```python
country_number = list(all_df.country.unique())
```

```python
def oh_encoding_country(df_origin):
    df = df_origin.copy()
    for country in country_number:
        df[f"country_{country}"] = list(np.where(np.array(df["country"]) == country, 1, 0))
    return df
```

```python
all_df_oh = oh_encoding_country(all_df)
all_df_oh
```

```python
category1_number = list(all_df_oh.category1.unique())
```

```python
def oh_encoding_category1(df_origin):
    df = df_origin.copy()
    for category1 in category1_number:
        df[f"category1_{category1}"] = list(np.where(np.array(df["category1"]) == category1, 1, 0))
    return df
```

```python
all_df_oh = oh_encoding_category1(all_df_oh)
all_df_oh.head(3)
```

```python
all_df_oh.to_csv('data/all_for_nn.csv', index=False)
```
